# ANÁLISE DOS DADOS OBTIDOS

> ### A seguir, com os dados tratados, começam algumas análises. O objetivo é comparar dados de aumento de casos do COVID-19 à venda de remédios do kit-covid, comparar os casos e as vendas a nível nacional e estadual, assim como analisar a venda de cada remédio por mês.

```python
# Importa as bibliotecas 

import pandas as pd
import numpy as np
```

```python
# Início da leitura dos milhões de arquivos com as bases do kit covid

file_list_2020=['data/tratados/202003.csv', 'data/tratados/202004.csv', 'data/tratados/202005.csv', 'data/tratados/202006.csv', 'data/tratados/202008.csv', 'data/tratados/202009.csv', 'data/tratados/202010.csv', 'data/tratados/202011.csv', 'data/tratados/202012.csv']
file_list_2021=['data/tratados/202101.csv', 'data/tratados/202102.csv', 'data/tratados/202103.csv', 'data/tratados/202104.csv', 'data/tratados/202105.csv', 'data/tratados/202106.csv', 'data/tratados/202107.csv', 'data/tratados/202108.csv', 'data/tratados/202109.csv', 'data/tratados/202110.csv', 'data/tratados/202111.csv', 'data/tratados/202112.csv']
file_list_2022=['data/tratados/202201.csv', 'data/tratados/202202.csv', 'data/tratados/202203.csv', 'data/tratados/202204.csv', 'data/tratados/202205.csv', 'data/tratados/202206.csv', 'data/tratados/202207.csv', 'data/tratados/202208.csv', 'data/tratados/202209.csv']
```

```python
# Inicio da concatenação dos dataframes

df_kitcovid_2020 = pd.DataFrame(pd.read_csv(file_list_2020[0]))
  
for i in range(1,len(file_list_2020)):
    data = pd.read_csv(file_list_2020[i])
    df = pd.DataFrame(data)
    df_kitcovid_2020 = pd.concat([df_kitcovid_2020,df],axis=0, ignore_index = True)
```

```python
df_kitcovid_2020.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
df_kitcovid_2020
```

```python
df_kitcovid_2021 = pd.DataFrame(pd.read_csv(file_list_2021[0]))
  
for i in range(1,len(file_list_2021)):
    data = pd.read_csv(file_list_2021[i])
    df = pd.DataFrame(data)
    df_kitcovid_2021 = pd.concat([df_kitcovid_2021,df],axis=0, ignore_index = True)
```

```python
df_kitcovid_2021.drop(["Unnamed: 0"], axis = 1, inplace = True)
df_kitcovid_2021
```

```python
df_kitcovid_2022 = pd.DataFrame(pd.read_csv(file_list_2022[0]))
  
for i in range(1,len(file_list_2022)):
    data = pd.read_csv(file_list_2022[i])
    df = pd.DataFrame(data)
    df_kitcovid_2022 = pd.concat([df_kitcovid_2022,df],axis=0, ignore_index = True)
```

```python
df_kitcovid_2022.drop(["Unnamed: 0"], axis = 1, inplace = True)
df_kitcovid_2022
```

```python
# Junção dos dataframes de cada ano para formar o total

df_kitcovid_total = pd.concat([df_kitcovid_2020, df_kitcovid_2021], axis = 0, ignore_index = True)
df_kitcovid_total
```

```python
df_kitcovid_total = pd.concat([df_kitcovid_total, df_kitcovid_2022], axis = 0, ignore_index = True)
df_kitcovid_total
```

> ## Com isso, temos os dados de venda de kitcovid de entre Março de 2020 a Setembro de 2022

```python
df = pd.read_csv('data/raw/cases-brazil-cities-time_changesOnly.csv')
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_filtrado = df.query("state == 'TOTAL'")[['date', 'newCases', 'totalCases', 'deaths', 'newDeaths']]
df_filtrado
```

```python
df_filtrado.rename(columns = {'date':'DATA', 'newCases': 'NOVOS_CASOS', 'totalCases': 'CASOS_TOTAIS', 'deaths':'MORTES', 'newDeaths': 'NOVAS_MORTES'}, inplace = True)
df_filtrado
```

```python
df_covid_novoscasos_mortes = df_filtrado
```

```python
df_covid_novoscasos = df_filtrado.groupby(df_filtrado['DATA'].dt.strftime('%y-%m-%d'))['NOVOS_CASOS'].sum().reset_index()
df_covid_novoscasos
```

```python
selecao = df_covid_novoscasos['DATA'] < '22-10-01'
df_covid_novoscasos2 = df_covid_novoscasos[selecao]
df_covid_novoscasos2
```
